- 이미지 빈자리 수 세기  
detect 후 도출된 txt 파일로 해당 이미지의 빈자리 수 를 알 수 있습니다.

```python
all = 0
cnt = 0

with open("/content/detected/images/labels/frame0068.txt", "r") as f:
    while True:
        line = f.readline()
        if not line: # 파일 읽기가 종료된 경우
            break
        if line.strip()[0]=="0":
          cnt += 1
        all += 1

print("빈 자리 수 : "+str(cnt))
print("전체 자리 수 : "+str(all))
```

```python
# 영상 끝에서부터 10초동안 빈자리 수 세기

import cv2
import os

filepath = '테스트 영상 경로(.mp4)'
video = cv2.VideoCapture(filepath)

fps = video.get(cv2.CAP_PROP_FPS) # 초당 프레임 수

path = "테스트 영상 레이블 폴더 경로(~/labels)"
files = os.listdir(path)

first = []
second = []

for i in files:
  if i[-6]=="_":
    first.append(i)
  else:
    second.append(i)

first.sort()
second.sort()
sorted_files = first+second
sorted_files

sum = 0
for i in range(len(sorted_files[-(10*int(fps)):])):
  cnt = 0
  with open(path + "/" + sorted_files[-(10*int(fps)):][i], "r") as f:
    while True:
        line = f.readline()
        if not line: # 파일 읽기가 종료된 경우
            break
        if line.strip()[0]=="0":
          cnt += 1
    sum += cnt

print("빈자리 수 :",int(sum/(int(fps)*10)))
```
