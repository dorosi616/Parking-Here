```python
# ÏòÅÏÉÅ ÎÅùÏóêÏÑúÎ∂ÄÌÑ∞ 10Ï¥àÎèôÏïà ÎπàÏûêÎ¶¨ Ïàò ÏÑ∏Í∏∞

import cv2
import os

filepath = 'ÌÖåÏä§Ìä∏ ÏòÅÏÉÅ Í≤ΩÎ°ú(.mp4)'
video = cv2.VideoCapture(filepath)

fps = video.get(cv2.CAP_PROP_FPS) # Ï¥àÎãπ ÌîÑÎ†àÏûÑ Ïàò

path = "ÌÖåÏä§Ìä∏ ÏòÅÏÉÅ Î†àÏù¥Î∏î Ìè¥Îçî Í≤ΩÎ°ú(~/labels)"
files = os.listdir(path)

first = []
second = []

for i in files:
  if i[-6]=="_":
    first.append(i)
  else:
    second.append(i)

first.sort()
second.sort()
sorted_files = first+second
sorted_files

sum = 0
for i in range(len(sorted_files[-(10*int(fps)):])):
  cnt = 0
  with open(path + "/" + sorted_files[-(10*int(fps)):][i], "r") as f:
    while True:
        line = f.readline()
        if not line: # ÌååÏùº ÏùΩÍ∏∞Í∞Ä Ï¢ÖÎ£åÎêú Í≤ΩÏö∞
            break
        if line.strip()[0]=="0":
          cnt += 1
    sum += cnt

print("ÎπàÏûêÎ¶¨ Ïàò :",int(sum/(int(fps)*10)))
```
